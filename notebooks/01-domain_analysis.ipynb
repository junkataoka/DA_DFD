```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
x_cwru = np.load("/data/home/jkataok1/DA_DFD/data/processed/CWRU/0.npz")["x"]
x_ims = np.load("/data/home/jkataok1/DA_DFD/data/processed/IMS/0.npz")["x"]
```

```python
scaler = MinMaxScaler()
x_ims_normalized = scaler.fit_transform(x_ims.reshape(-1, 2048))
x_cwru_normalized = scaler.fit_transform(x_cwru.reshape(-1, 2048))
plt.plot(x_ims_normalized[0, :100])
plt.plot(x_cwru_normalized[0, :100])
```

```python
# compute Similarity between two domains
lm = LinearRegression().fit(x_ims_normalized[:1709, :1000], x_cwru_normalized[:1709, :1000])
# Compute accuracy of the model
lm.score(x_ims_normalized[:1709, :1000], x_cwru_normalized[:1709, :1000])
```
